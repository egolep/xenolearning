### Per un attimo, non preoccupiamoci del codice e della sintassi

Questa riga di codice si limita a prendere un modello di ML appositamente creato e "istruito" (nel prossimo episodio vedremo il codice usato per queste operazioni) e chiedergli di "predirre" il valore della nostra funzione iniziale per il valore 1000

Molto vicino ma non esatto, eh?  
Esattamente. E questo è il primo grande punto da tenere a mente, sia quando vediamo un incredibile risultato ottenuto tramite l'AI, sia quando vediamo un fallimento epico: il ML non è uno strumento che lavora cercando di ottenere un risultato ESATTO, bensì uno strumento statistico, ovvero che cerca di trovare una soluzione APPROSSIMATA ma il più simile e vicina possibile alla realtà.

Questo approccio ha diversi vantaggi:
- Non è necessario conoscere una soluzione (esatta o meno) per un dato problema. Basterà sviluppare un modello che approssimi sufficientemente bene la soluzione
- Anche se si conosce una soluzione esatta, a volte calcolarla effettivamente è impossibile o il problema è intrinsecamente intrattabile[<sup>1<sup>](#note1). Approssimare questa soluzione risulterebbe sicuramente molto più veloce
- [Tutto è approssimabile][*], quindi in linea di principio è possibile utilizzare l'AI per risolvere (sempre tramite approssimazione) qualsiasi problema

  

[*]: https://en.wikipedia.org/wiki/Universal_approximation_theorem

Un altro vantaggio, se non il più grande, è stato già evidenziato in precedenza: siccome non stiamo ricercando una soluzione specifica di un problema, ma una sua approssimazione, è possibile usare lo stesso algoritmo per risolvere diversi problemi, dovendo "solamente" cambiare il processo di "insegnamento".

Ovviamente, come sempre nella vita, ci sono anche degli svantaggi:
- Il primo, già visto: si può ottenere una approssimazione anche molto accurata, ma è impossibile ottenere una soluzione esatta
- In base a quanto è complesso il problema da risolvere, la fase di "addestramento" può essere anche parecchio dispendiosa e complicata
- Come conseguenza, è possibile che approssimare la soluzione di un dato problema sia talmente complesso e richieda talmente tante risorse da risultare di fatto infattibile

### Se vi sembra che vantaggi e benefici abbiano un che di speculare... beh, è così

Infatti, in base al caso, può essere più vantaggioso utilizzare un forma approssimata e/o il ML, oppure cercare di trovare e utilizzare una soluzione esatta.

Come sempre, la realtà non è bianca o nera e quindi, soprattutto negli ultimi anni, stanno emergendo tante forme intermedie per cercare di mantenere i vantaggi di entrambi gli approcci, mitigando le rispettive debolezze.
